```python
continuous = [
    'Age',                     # Integer (e.g., 22, 30)
    'High_School_GPA',         # Float (e.g., 3.2, 4.0)
    'SAT_Score',               # Integer (e.g., 1200, 1600)
    'University_Ranking',      # Integer (e.g., 1, 50) #if treated as ordinal*
    'University_GPA',          # Float (e.g., 2.8, 3.9)
    'Internships_Completed',   # Integer (e.g., 0, 5)
    'Projects_Completed',      # Integer (e.g., 2, 10)
    'Certifications',          # Integer (e.g., 0, 3)
    'Soft_Skills_Score',       # Integer (e.g., 5, 10)
    'Networking_Score',        # Integer (e.g., 4, 9)
    'Job_Offers',              # Integer (e.g., 1, 4)
    'Starting_Salary',         # Float (e.g., 50000.0, 85000.0)
    'Career_Satisfaction',     # Integer (e.g., 3, 9) #if treated as ordinal*
    'Years_to_Promotion',      # Integer (e.g., 1, 5)
    'Work_Life_Balance'        # Integer (e.g., 2, 8) #if treated as ordinal*
]
print(continuous)
```

```python
from scipy import stats
```

```python
#continious vs continious
def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.1:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
two_sample(df["High_School_GPA"],df["University_Ranking"])
```

```python
from itertools import combinations

continuous_cols = [
    'Age',
    'High_School_GPA',
    'SAT_Score',
    'University_Ranking',
    'University_GPA',
    'Internships_Completed',
    'Projects_Completed',
    'Certifications',
    'Soft_Skills_Score',
    'Networking_Score',
    'Job_Offers',
    'Starting_Salary',
    'Career_Satisfaction',
    'Years_to_Promotion',
    'Work_Life_Balance'
]

for col1, col2 in combinations(continuous_cols, 2):
    result = two_sample(df[col1], df[col2])
    print(f"Two sample test between '{col1}' and '{col2}': {result}")
```

```python
#cat vs cat
def chisquare_cat_vs_cat(d1,d2):
    return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.1 else False
```

```python
from itertools import combinations

categorical_cols = [
    'Gender',
    'Field_of_Study',
    'Current_Job_Level',
    'Entrepreneurship'
]

for col1, col2 in combinations(categorical_cols, 2):
    result = chisquare_cat_vs_cat(df[col1], df[col2])
    print(f"Chi-square test between '{col1}' and '{col2}': {result}")
```

```python
#continious(d1) vs cat(d2)
def annova_test(d1,d2):
    group= df[d2].unique()
    data = {}
    for i in group:
        data[i]=df[d1][df[d2]==i]
    f_value,p_value=stats.f_oneway(*[i for i in data.values()])
    return False if p_value < 0.1 else True
```

```python
annova_test("Entrepreneurship","Job_Offers")
```

```python
final = {}
# Exclude datetime columns from the list of columns to iterate through
cols_to_analyze = [col for col in df.columns if col not in ['Entrepreneurship', 'Job_Offers']]

for i in cols_to_analyze:
    final[i]={}
    for j in cols_to_analyze:
        if (i in continuous) and (j in continuous):
            result=two_sample(df[i],df[j])
        elif (i in continuous) and (j in categorical):
            result=annova_test(i,j)
        elif (i in categorical) and (j in continuous):
            result=annova_test(j,i)
        elif (i in categorical) and (j in categorical):
            result=chisquare_cat_vs_cat(df[i],df[j])
        else:
            # If the columns are of mixed types not covered by the above or one is a date, skip the test
            continue

        if result:
            final[i][j]=1
        else:
            final[i][j]=0
```
